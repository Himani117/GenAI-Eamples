```python
from agents import Agent, Runner, trace
from dotenv import load_dotenv
load_dotenv()
```

```python
joke_agent = Agent(
    name="Joke Agent",
    instructions="You are a joke-telling agent. You are provided with a topic and you must tell a joke related to that topic.",
    )

language_agent = Agent(
    name="Language Agent",
    instructions="You are a language translation agent. You are provided with a sentence and you must translate it to different language.",
    )

with trace("Joke Translation Workflow"):
    joke_result = await Runner.run(joke_agent, "Booger")
    print(joke_result.final_output)
    transalated_joke = await Runner.run(language_agent, f"Translate the following joke to Hindi: {joke_result.final_output}")
    print(transalated_joke.final_output)
```
